```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
X = db.iloc[:,4].values
db.iloc[:,4] = labelencoder.fit_transform(X)
X = X.reshape(-1,1)
```

```python
X= db[['gender','age','hypertension','heart_disease','smoking_history','bmi','HbA1c_level','blood_glucose_level',]]
Y= db[['diabetes']]

from sklearn.model_selection import train_test_split

# split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X_train_min= mms.fit_transform(X_train)
X_test_min = mms.transform(X_test)
from sklearn.preprocessing import StandardScaler
independent_scaler = StandardScaler()
X_train_norm = independent_scaler.fit_transform(X_train)
X_test_norm = independent_scaler.transform(X_test)
from sklearn.decomposition import PCA
pca = PCA(n_components =2)              # Convert All the columns into two components/columns
X_train_pca = pca.fit_transform(X_train_norm)
X_test_pca =pca.transform(X_train_norm)
X_train_pca = pca.fit_transform(X_test_norm)
X_test_pca =pca.transform(X_test_norm)
pca.explained_variance_ratio_
```

```python
from sklearn.decomposition import PCA
from sklearn.svm import SVC

# standardize the data and perform dimensionality reduction
X_scaled = (X_train - X_train.mean()) / X_train.std()
pca = PCA(n_components=8)
X_train_pca = pca.fit_transform(X_scaled)

# fit the model
svc = SVC()
svc.fit(X_train_pca, Y_train.values.ravel())

# reduce the dimensionality of the testing data
X_test_scaled = (X_test - X_train.mean()) / X_train.std()
X_test_pca = pca.transform(X_test_scaled)

# make predictions on the testing data
predictions = svc.predict(X_test_pca)
```

```python
#Logistic Regression MinMax
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report ,confusion_matrix ,ConfusionMatrixDisplay
lgr = LogisticRegression()
lgr.fit(X_train_min , Y_train)

prediction = lgr.predict(X_test_min)
from sklearn import metrics
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test,prediction)
print(accuracy)

target_names = ['diabitic','non diabitic']
print(classification_report(Y_test, prediction.reshape(-1,1), target_names=target_names))


cm = confusion_matrix(Y_test, prediction.reshape(-1,1), normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
fig, ax = plt.subplots(figsize=(8,6))
disp.plot(ax=ax,cmap='Blues')
```

```python
#Logistic Regression PCA
from sklearn.linear_model import LogisticRegression
lgr = LogisticRegression()
lgr.fit(X_train_pca , Y_train)

prediction = lgr.predict(X_test_pca)
from sklearn import metrics
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test,prediction)
print(accuracy)

target_names = ['diabitic','non diabitic']
print(classification_report(Y_test, prediction.reshape(-1,1), target_names=target_names))

cm = confusion_matrix(Y_test, prediction.reshape(-1,1), normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
fig, ax = plt.subplots(figsize=(8,6))
disp.plot(ax=ax,cmap='Blues')
```
